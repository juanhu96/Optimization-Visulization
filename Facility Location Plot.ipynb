```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import random
from math import *
```

```python
cities = ["city_1", "city_2", "city_3", "city_4", "city_5", "city_6", "city_7", "city_8", "city_9", "city_10"]

cities_coord = {"city_1":{"x": 5,"y": 41},
              "city_2":{"x": 20,"y": 10},
              "city_3":{"x": 44,"y": 48},
              "city_4":{"x": 60,"y": 58},
              "city_5":{"x": 100,"y": 4},
              "city_6":{"x": 138,"y": 80},
              "city_7":{"x": 150,"y": 40},
              "city_8":{"x": 170,"y": 18},
              "city_9":{"x": 182,"y": 2},
              "city_10":{"x": 190,"y": 56}
               }
```

```python
plt.plot([5,20,44,60,100,138,150,170,182,190],[41,10,48,58,4,80,40,18,2,56], 'ro')
plt.plot([114.1], [35.2], 'bo')
```

```python
%matplotlib notebook
def dist(x,y):
    total = 0
    for city in cities:
        total = total + np.sqrt((x - cities_coord[city]["x"])**2 + (y - cities_coord[city]["y"])**2)
    return total
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(0.0, 190.0, 1.0)
y = np.arange(0.0, 80.0, 1.0)
X, Y = np.meshgrid(x, y)
Z = dist(X, Y)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,vmin =0, vmax=np.nanmax(Z), cmap='viridis',
                       linewidth=0, antialiased=True)
ax.contourf(X, Y, Z, zdir='z', offset=-100, cmap='viridis', antialiased=True)
# set axis
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')


plt.show()
```
